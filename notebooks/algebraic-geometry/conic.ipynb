# 二次曲線

```sage
x, y = AffineSpace(2, QQ, 'xy').gens()
```

```sage
C1 = Curve(y - x^2)
```

```sage
plot(C1)
```

```sage
C2 = Curve(x^2 / 4 + y^2 / 9 - 1)
```

```sage
plot(C2)
```

```sage
C3 = Curve(x - y^2)
```

```sage
plot(C3)
```

```sage
C4 = Curve(x^2 - y^2 - 4)
```

```sage
plot(C4)
```
