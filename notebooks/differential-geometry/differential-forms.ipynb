```python
M = Manifold(2, 'M')
U = M.open_subset('U') ; V = M.open_subset('V')
M.declare_union(U,V)   # M is the union of U and V
c_xy.<x,y> = U.chart() ; c_uv.<u,v> = V.chart()
```

```python
xy_to_uv = c_xy.transition_map(
  c_uv,
  (x+y, x-y),
  intersection_name='W',
  restrictions1= x>0, restrictions2= u+v>0
)
uv_to_xy = xy_to_uv.inverse()
```

```python
W = U.intersection(V)
eU = c_xy.frame() ; eV = c_uv.frame()
```

```python
a = M.diff_form(1, name='a'); a
```

```python
a.parent()
```

```python
a.degree()
```

```python
# setting components
a[eU,:] = [-y, x]
a.add_comp_by_continuation(eV, W, c_uv)
```
