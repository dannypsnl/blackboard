```python
M = Manifold(4, 'M', structure='Lorentzian', start_index=1)
X.<t,x,y,z> = M.chart()
g = M.metric()
g[1,1] = -1
g[2,2] = 1
g[3,3] = 1
g[4,4] = 1
g.display()
```

```python
nab = g.connection()
nab.set_immutable()
nab
```

```python
TM = M.tangent_bundle()
p = TM.characteristic_cohomology_class('Pontryagin')
p_form = p.get_form(nab)
p_form.display_expansion()
```
